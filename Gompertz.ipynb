```python
for i in range(0, longitud):
    if gompertz_Colombia.loc[i, 'ID de caso'] == 0:
        pass
    else:
        gompertz_Colombia.loc[i, 'ID de caso'] = 1
```

```python
df = gompertz_Colombia.groupby('Fecha de notificación')['ID de caso'].sum().to_frame()
```

```python
df
```

```python
df.reset_index(inplace=True)
```

```python
df['Acumulados']= 0
```

```python
df['Día número']= 0
```

```python
df['Penetración real']= 0
```

```python
df
```

```python
df.size
```

```python
longitud = 164
```

```python
df.loc[0, 'Acumulados']= df.loc[0, 'ID de caso']
for i in range(1, longitud):
    df.loc[i,'Acumulados'] = df.loc[i, 'ID de caso'] + df.loc[i-1, 'Acumulados']
    
```

```python
df.loc[0, 'Día número'] = 1
for i in range(1, longitud):
    df.loc[i,'Día número'] = i+1
    df.loc[i,'Penetración real'] = df.loc[i, 'Acumulados']/ df.loc[longitud-1, 'Acumulados']
    
```
