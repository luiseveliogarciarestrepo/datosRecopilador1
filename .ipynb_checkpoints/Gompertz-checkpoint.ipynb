```python
gompertzbyFecha= funcion_gompertz.groupby(['Fecha de notificación'])[['Departamento o Distrito ']].count()
```

```python
gompertzbyFecha.columns
```

```python
trabajo = gompertzbyFecha.rename({'Departamento o Distrito ':'Casos Notificados diarios'}, axis = 'columns')
```

```python
trabajo=trabajo.reset_index()
```

```python
trabajo['Acumulados']= 0
trabajo['Día número']=0
trabajo['Penetración real']=0
trabajo
```

```python
trabajo.index

```

```python
longitud2 = trabajo['Casos Notificados diarios'].size
longitud2
```

```python
trabajo.loc[0, 'Acumulados']= trabajo.loc[0, 'Casos Notificados diarios']
for i in range(1, longitud2):
    trabajo.loc[i,'Acumulados'] = trabajo.loc[i, 'Casos Notificados diarios'] + trabajo.loc[i-1, 'Acumulados']
    
```

```python
trabajo
```

```python
trabajo.loc[longitud2-1, 'Acumulados']
```

```python
trabajo.loc[0, 'Día número'] = 1
for i in range(1, longitud2):
    trabajo.loc[i,'Penetración real'] = trabajo.loc[i, 'Acumulados']/ trabajo.loc[longitud2-1, 'Acumulados']
    trabajo.loc[i,'Día número'] = i+1
```
