```python
import pandas as pd
```

```python

df = pd.read_csv('DaneMunic2018_2023.csv')
```

```python
df
```

```python
filtro = df['Género'] == 'AMBOS SEXOS'
df2 = df.loc[filtro, ['Dpto', 'Municipio', 'Género', '2018', '2019', '2020', '2021', '2022',
       '2023']]
```

```python
df2.columns
```

```python
municipios=['Manizales', 'Barranquilla', 'Pereira', 'Armenia', 'Neira']
dptos = ['Caldas', 'Risaralda', 'Quindio']
dfgrupomunic=  df2.groupby(['Dpto', 'Municipio'])[['Género', '2020', '2021', '2022',
       '2023']].sum().loc[dptos]
dfgrupodptos=  df2.groupby(['Dpto'])[['Género', '2020', '2021', '2022',
       '2023']].sum().loc[dptos]
```
