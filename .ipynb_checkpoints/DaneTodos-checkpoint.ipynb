```python
import pandas as pd
```

```python
df = pd.read_csv('CovidColombia.csv')
df_col_munic = pd.read_csv('DaneMunic2018_2023.csv')
```

```python
df_agrup = df.groupby('Departamento o Distrito ')
```

```python
df_agrup
```

```python
df_agrup.get_group('Caldas')
```

```python
df_agrup['Departamento o Distrito '].value_counts()
```

```python
df_agrup['Estado'].apply(lambda x:x.str.contains('Fallecido').sum())
```

```python
casos_dpto= df['Departamento o Distrito '].value_counts()
```

```python
fallecidos = df_agrup['Estado'].apply(lambda x:x.str.contains('Fallecido').sum())
```

```python
letalidad_df = pd.concat([casos_dpto, fallecidos], axis='columns', sort=False)
```

```python
letalidad_df
```

```python
letalidad_df.rename(columns= {'Departamento o Distrito ':'Número de casos', 'Estado':'Fallecidos'}, inplace=True)
```

```python
letalidad_df
```

```python
letalidad_df['Tasa de letalidad %'] = (letalidad_df['Fallecidos']/letalidad_df['Número de casos'])*100
```

```python
letalidad_df.sort_values(by='Tasa de letalidad %', ascending=False, inplace=True)
```

```python
letalidad_df.loc['Caldas']
```
