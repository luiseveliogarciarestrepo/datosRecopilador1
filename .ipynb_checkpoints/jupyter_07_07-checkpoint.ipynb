```python
fallecidos =ciudades['Estado'].apply(lambda x: x.str.contains('Fallecido').sum())
```

```python
fallecidos_dep = departamentos['Estado'].apply(lambda x:x.str.contains('Fallecido').sum())
```

```python
fallecidos
```

```python
fallecidos_dep.sort_values()
```

```python
casos_ciudad = df['Ciudad de ubicación'].value_counts()
```

```python
casos_departamento = df['Departamento o Distrito '].value_counts()
```

```python
casos_ciudad
```

```python
casos_departamento
```

```python
 concatenada= pd.concat([casos_ciudad, fallecidos], axis= 'columns', sort=False)
```

```python
concat_dptos= pd.concat([casos_departamento, fallecidos_dep], axis='columns')
```

```python
concatenada
```

```python
concat_dptos
```

```python
concatenada.rename(columns={'Ciudad de ubicación':'Número casos', 'Estado':'Fallecidos'}, inplace=True)
```

```python
concat_dptos.rename(columns={'Departamento o Distrito ': 'Número casos', 'Estado':'Fallecidos'},inplace=True)
```

```python
concatenada
```

```python
concatenada['Porcentaje'] = (concatenada['Fallecidos']/concatenada['Número casos']) *100
```

```python
concat_dptos['Porcentaje']= (concat_dptos['Fallecidos']/concat_dptos['Número casos']) *100
```

```python
concatenada
```

```python
concatenada.sort_values(by='Porcentaje', ascending=True, inplace=True)
```

```python
concat_dptos.sort_values(by='Porcentaje', ascending= False, inplace=True)
```
