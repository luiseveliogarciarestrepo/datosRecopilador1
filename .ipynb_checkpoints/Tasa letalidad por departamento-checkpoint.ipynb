```python
import pandas as pd
```

```python
df = pd.read_csv('CovidColombia.csv')
df_col = pd.read_csv('DaneColDpto.csv')
```

```python
pd.set_option('display.max_rows',1000)
```

```python
df_agrup = df.groupby('Departamento o Distrito ')
```

```python
df_agrup.get_group('Caldas')
```

```python
df_agrup['Estado'].apply(lambda x:x.str.contains('Fallecido').sum())
```

```python
casos_departamento= df['Departamento o Distrito '].value_counts()
```

```python
fallecidos = df_agrup['Estado'].apply(lambda x:x.str.contains('Fallecido').sum())
```

```python
letalidad_df = pd.concat([casos_departamento, fallecidos], axis='columns', sort=False)
```

```python
letalidad_df
```

```python
letalidad_df.rename(columns= {'Departamento o Distrito ':'Número de casos', 'Estado':'Fallecidos'}, inplace=True)
```

```python
letalidad_df
```

```python
poblacionxDpto= df_col.set_index('Dpto')
```

```python
indicadores = pd.concat([letalidad_df, poblacionxDpto], axis='columns', sort=False)
```

```python
indicadores
```

```python
indicadores.rename(columns={'2020':'Población 2020'}, inplace=True)
```

```python
indicadores
```

```python
indicadores['Tasa de Letalidad %'] = (indicadores['Fallecidos']/indicadores['Número de casos'])*100
indicadores['Tasa de mortalidad por 100.000'] = (indicadores['Fallecidos']/indicadores['Población 2020'])*100000
```

#### Tasas de letalidad y mortalidad CoVid-19 por Departamentos y País
###### Datos al 8 de julio 2020.  Fuentes:  INS, Dane
###### Luis Evelio García Restrepo Md. MAS. EF.

```python
indicadores.sort_values(by='Tasa de mortalidad por 100.000',ascending=False)
```

```python
TasaLetalidadColombia =  (indicadores['Fallecidos'].sum()/indicadores['Número de casos'].sum())*100
TasaLetalidadColombia # En porcentaje
```

```python
TasaMortalidadColombia = (indicadores['Fallecidos'].sum()/indicadores['Población 2020'].loc['Total'])*1000000
TasaMortalidadColombia # Por millón de habitantes
```
