```python
df_fechas['ID de caso']=0
```

```python
for i in range(1,longitud):
    df_fechas.loc[(i)] = [(np.datetime64(df_fechas.loc[i-1, 'Fecha de notificación']) + np.timedelta64(1,'D')),0]
    
```

```python
df_fechas
```

```python
gompertz_Col=gompertz.append(df_fechas).sort_values(by=['Fecha de notificación'], ignore_index=True)
```

```python
gompertz_Col
```

```python
longitud = gompertz_Col['ID de caso'].size
longitud
```

```python
for i in range(0, longitud):
    if gompertz_Col.loc[i, 'ID de caso'] == 0:
        pass
    else:
        gompertz_Col.loc[i, 'ID de caso'] = 1
```

```python
#pd.to_datetime(gompertz_Col.loc[237958,'Fecha de notificación'], format='%Y-%m-%d')
#gompertz_Col.loc[237958, 'Fecha de notificación']
```

```python
gompertz_Col['Acumulados']= 0
gompertz_Col['Día número']=0
gompertz_Col['Penetración real']=0
gompertz_Col
```

```python
gompertz_Col.shape
```

```python
gompertz_Col.loc[0, 'Acumulados']= gompertz_Col.loc[0, 'ID de caso']
for i in range(1, longitud):
    gompertz_Col.loc[i,'Acumulados'] = gompertz_Col.loc[i, 'ID de caso'] + gompertz_Col.loc[i-1, 'Acumulados']
```

```python
gompertz_Col
```

```python
gompertz_Colombia = gompertz_Col.groupby('Fecha de notificación')[['ID de caso', 'Ciudad de ubicación',
       'Departamento o Distrito ', 'Acumulados', 'Día número',
       'Penetración real']].max()
```

```python
gompertz_Colombia
```

```python
gompertz_Colombia.reset_index(inplace=True)
```

```python
longitud = gompertz_Colombia['ID de caso'].size
longitud
```

```python
gompertz_Colombia.columns
```

```python
gompertz_Colombia.loc[0, 'Día número'] = 1
for i in range(1, longitud):
    gompertz_Colombia.loc[i,'Día número'] = i+1
    gompertz_Colombia.loc[i,'Penetración real'] = gompertz_Colombia.loc[i, 'Acumulados']/ gompertz_Colombia.loc[longitud-1, 'Acumulados']
    
```
