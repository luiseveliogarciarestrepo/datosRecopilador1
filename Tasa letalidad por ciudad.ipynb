```python
df_agrup = df.groupby('Ciudad de ubicación')
```

```python
casos_ciudad= df['Ciudad de ubicación'].value_counts()
```

```python
fallecidos = df_agrup['Estado'].apply(lambda x:x.str.contains('Fallecido').sum())
```

```python
letalidad_df = pd.concat([casos_ciudad, fallecidos], axis='columns', sort=False)
```

```python
pd.set_option('display.max_rows',100)
```

```python
letalidad_df
```

```python
letalidad_df.rename(columns= {'Ciudad de ubicación':'Número de casos', 'Estado':'Fallecidos'}, inplace=True)
```

```python
letalidad_df
```

```python
letalidad_df['Tasa de letalidad %'] = (letalidad_df['Fallecidos']/letalidad_df['Número de casos'])*100
```

```python
letalidad_df.sort_values(by='Tasa de letalidad %', ascending=False, inplace=True)
```
